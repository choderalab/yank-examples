```python
from netCDF4 import Dataset
filename = 'output/experiments/freesolvhydrationbinders0/solvent1.nc'
print(os.path.exists(filename))
ncfile = Dataset(filename, 'r')
```

```python
logZ = ncfile.groups['online_analysis'].variables['logZ_history']
n_iterations, n_states = logZ.shape
```

```python
fig = plt.figure(figsize=(10,5));
plt.plot(np.transpose(np.transpose(logZ) - logZ[:,0]),'.');
plt.xlabel('iteration');
plt.ylabel('logZ / kT');
```

```python
states = ncfile.variables['states']
n_iterations, n_replicas = states.shape
```

```python
fig = plt.figure(figsize=(10,5));
plt.plot(states,'.');
plt.xlabel('iteration');
plt.ylabel('thermodynamic state');
plt.axis([0, n_iterations, 0, n_states]);
```
